```python
#Dependencies
import pandas as pd
from glob import glob
import numpy as np
```

```python
# list all files that match a pattern and sort results
stock_files = sorted(glob('raw_data/data_*.csv'))
stock_files
```

```python
# Use a generator expression to read the files, assign(), to create a new column, and concat Dataframes 
df = pd.concat((pd.read_csv(file).assign(filename = file)
          for file in stock_files), ignore_index = True)
df.head()
```

```python
#rename index head
df = df.rename(columns={"Province_State": "state"})
df
```

```python
#Drop the only column for rows that contains another data type  besides Nan
df1=df.drop(columns= ['filename'])
df1
```

```python
#Drop all rows with only Nan values
df2=df1.dropna(how='all')
df2
```

```python
#Replace all Nan cell value to Zero
df3 = df2.replace(np.nan, '0', regex=True)
df3
```

```python
#Identify the data types before summimg up the values
df3.dtypes
```

```python
#Convert the data type of those columns that contains numberic values from object to float64
df3=df3.astype({'Recovered':'float64'})
df3=df3.astype({'Active':'float64'})
df3=df3.astype({'People_Tested':'float64'})
df3=df3.astype({'People_Hospitalized':'float64'})
```

```python
#Group the data by state and sum up the values
df4=df3.groupby(['state']).sum()
df4
```

```python
# Create CSV file
df4.to_csv('john_hopkins.csv')
```

```python
#Create Database connection
from sqlalchemy import create_engine
connection_string = "postgres:postgres@localhost:5432/etl_db"
engine = create_engine(f'postgresql://{connection_string}')
```

```python
# Confirm tables
engine.table_names()
```

```python
#Load dataframes to database
df4.to_sql(name='john_hopkins_data', con=engine, if_exists='append', index=True)
```

```python
#Get the teammate data
cdc_file = "raw_data/cdc_may_group_data.csv"
cdc_file_df = pd.read_csv(cdc_file)
cdc_file_df.head()
```

```python
#Load dataframes to database
cdc_file_df.to_sql(name='cdc_covid_data', con=engine, if_exists='append', index=True)
```

```python
#Merge the two dataframes
combined_data = pd.merge(df4,cdc_file_df, on='state')
combined_data.set_index("state", inplace=True)
combined_data
```

```python
# Save Data to csv
combined_data.to_csv("combined_data.csv")
```
